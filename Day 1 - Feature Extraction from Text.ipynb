```python
import pandas as pd
from textblob import TextBlob

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
df = pd.read_csv(r"dataset/AvengersEndgame_tweets.csv" , sep=',', encoding='cp1252')
```

```python
df.head(5)
```

```python
df.shape
```

```python
# in this file we are gonna only work on the text data
text = df[['text']]
```

```python
print(text.head(15).to_csv(sep='\n', index=False))
```

# Feature Extraction: Round 1

```python
# 1. Number of keyword
text['hashtags_count'] = text['text'].apply( lambda x : len( [ x for x in x.split() if x.startswith('#') ]) )

text[['text' , 'hashtags_count']].head(5)
```

```python
# 2. Number of Users Tagged
text['users_tagged'] = text['text'].apply( lambda x : len( [ x for x in x.split() if x.startswith('@') ]) )

text[['text' , 'users_tagged']].head(5)
```

```python
# 3. Number of numerical values 
# similar to counting number of hashtags
text['numeric'] = text['text'].apply( lambda x : len( [ x for x in x.split() if x.isdigit() ]) )

text[['text' , 'numeric']].head(10)
```

```python
# Number of UPPERCASE words
text['upper'] = text['text'].apply( lambda x : len( [ x for x in x.split() if x.isupper() ]) )
text[['text' , 'upper']].head(5)
```

# Text Cleaning

It's time to clear out the symbols, punctuations because they never contribute anything to the models



```python
# the [^\w\s] means remove everything, keep only words(w) and spaces(s)

text['text'] = text['text'].str.replace('[^\w\s]','')
text['text'].head()
```

```python
# remove \r\n from the tweets
text['text'] = text['text'].str.replace("\r\n","") 
text['text'].head(5)
```

```python
# remove links from the tweet
def remove_http(x):
    tweet = x[:-19]
    return tweet
        
text['text'] = text['text'].apply( lambda x: remove_http(x) )
text['text'].tail(15)
```

# Feature Extraction : Round 2

```python
#  Feature : Average word length 

def avg_word_length(t):
    words = t.split()
    return ( sum( len(word) for word in words ) / len(words))

text['avg_word_len'] = text['text'].apply( lambda x: avg_word_length(x) )
```

```python
text[['text' , 'avg_word_len']].head(5)
```

```python
# Feature extraction : number of words
text['word_count'] = text['text'].apply( lambda x : len(str(x).split(" ") ) )

text[['text' , 'word_count']].head(5)
```

```python
# sentiment score calculation
#sentiment score, polarity, subjectivity and intensity can be calculated
def sentiment(x):
    sentiment = TextBlob(x)
    return sentiment.sentiment.polarity

text['sentiment_score'] = text['text'].apply(sentiment)
text.head(5)
```

```python
# assign polarity score to the sentiments

sentiment_numeric = []

for score in text['sentiment_score']:
    # positive sentiments will have value "+1"
    if score > 0:
        sentiment_numeric.append("positive")
    # positive sentiments will have value "-1"
    elif score < 0:
        sentiment_numeric.append("negative")
    # positive sentiments will have value "0"
    elif score == 0:
        sentiment_numeric.append("neutral")

text['sentiment'] = sentiment_numeric
text.sentiment.value_counts()
```

```python
text[['text' , 'sentiment_score', 'sentiment']].tail(5)
```

```python
# Bag of Words
cv = CountVectorizer(stop_words='english')
text_cv = cv.fit_transform(text.text.values.astype('str'))
```

```python
# tf-idf, better than bag of words
#tfidf = TfidfVectorizer(sublinear_tf=True, min_df=50, norm='l2', encoding='latin-1')
tfidf = TfidfVectorizer(norm=None)
text_tfidf = tfidf.fit_transform(text.text).toarray()

```
